Constructor

```python
class Homosapien:
    def __init__(self, name, age,height,weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight
    def bmi(self):
        return self.weight / (self.height ** 2)
p1=Homosapien("Alice", 30, 1.65, 68)
print(p1.name)
p1.bmi()
```

Inheritance

```python
class animal:
    def eat(self):
        print("Animal is eating")
    def sleep(self):
        print("Animal is sleeping")
class Dog(animal):
    def bark(self):
        print("Dog is barking") 
class Cat(animal):
    def meow(self):
        print("Cat is meowing")

jackie = Dog()
steve = Cat()
jackie.eat()
steve.sleep()
jackie.bark()
```

MULTIPLE INHERITANCE

```python
class Father:
    def method1(self):
        print("Method 1 from Father")
class Mother:
    def method2(self):
        print("Method 2 from Mother")   
class Child(Father, Mother):
    def method3(self):
        print("Method 3 from Child")        
child = Child()
child.method1()
child.method2()
child.method3()
```

MULTILEVEL INHERITANCE

```python
class grandfather:
    def method1(self):
        print("Method 1 from Grandfather")
class Father(grandfather):
    def method2(self):
        print("Method 2 from Father")

class Child(Father):
    def method3(self):
        print("Method 3 from Child")        
child = Child()
child.method1()
child.method2()
child.method3()
```

HIERARCHICAL INHERITANCE

```python
class father:
    def method1(self):
        print("Method 1 from Father")
class son(father):
    def method2(self):
        print("Method 2 from Son")
class daughter(father):
    def method3(self):
        print("Method 3 from Daughter")
class grandson(son):
    def method4(self):
        print("Method 4 from Grandson")
class granddaughter(daughter):
    def method5(self):
        print("Method 5 from Granddaughter")
grandson1 = grandson()
granddaughter1 = granddaughter() 
print("this from grand son")    
print("---------------------\n")   
grandson1.method1()
grandson1.method2() 
print("\n---------------------\n")
print("this from grand daughter")
print("---------------------\n")
granddaughter1.method1()
granddaughter1.method3()
granddaughter1.method5()
```

HYBRID INHERITANCE

```python
class Family:
    def method1(self):
        print("Method 1 from Family")
class Father(Family):
    def method2(self):
        print("Method 2 from Father")
class Mother(Family):
    def method3(self):
        print("Method 3 from Mother")
class Child(Father, Mother):
    def method4(self):
        print("Method 4 from Child")    
child = Child()
child.method1()     
child.method2()
child.method3() 
child.method4()
```

Method Overridding

```python
class Animal:
  def __init__(self, name):
    self.name = name

    def sound(self):
        print("Animal makes a sound")
    def describe(self):
        print(f"{self.name} is an animal")
class Dog(Animal):
    def sound(self):
        print(f"{self.name} barks")
class Cat(Animal):
    def sound(self):
        print(f"{self.name} meows")
class Cow(Animal):
    def sound(self):
        print(f"{self.name} moos")
dog1 = Dog("Buddy")
cat1 = Cat("Whiskers")      
cow1 = Cow("Bessie")
dog1.sound()
cat1.sound()
cow1.sound()
```
