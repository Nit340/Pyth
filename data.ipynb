Methods

```python
fruits=["apple","banana","grapes","orange"]
fruits.append("kiwi")
print(fruits)   
fruits.insert(1,"mango")
print(fruits)   
fruits.remove("banana")
print(fruits)   
fruits.pop()
print(fruits)
fruits.pop(1)   
print(fruits)   
fruits.sort(reverse=True)
print(fruits)
fruits.reverse()
print(fruits)   
fruits.count("apple")
fruits.index("grapes")
fruits.clear()
print(fruits)
```
