If

```python
menu = ["idly","dosa","vada","poori"]
if "idly" in menu:
    print("idly is available")
elif "dosa" in menu:
    print("dosa is available")
elif "vada" in menu:
    print("vada is available")
else:
    print("poori is available")
```

match case

```python
country = input("Enter your country name :")
match country:
    case "India":
        print("Namaste")
    case "USA":
        print("US Hello")
    case "UK":  
        print("UK Hello")
    case "Japan":
        print("Japan Hello")
    case _:
        print("Hello")
```
